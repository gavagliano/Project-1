```python
import requests
import json
import pandas as pd
```

```python
url = "https://api.foursquare.com/v3/places/search?categories=13034%2C13035&exclude_all_chains=true&fields=name%2Chours%2Cgeocodes%2Clocation%2Cdate_closed%2Cdescription%2Cfeatures&open_at=1T0000&near=New%20York%2C%20NY&limit=50"

headers = {
    "accept": "application/json",
    "Authorization": "fsq3+i9wQXUHtAvYAUtJzPszGzJjz69anbZluHb3N2tdusM="
}

response = requests.get(url, headers=headers)

print(response.text)
data = response.json()
```

```python
for case in data["results"]:
    print(case['name'])
```

```python
list = []
for case in data["results"]:
    dict = {}
    dict['name'] = case['name']
    dict['latitude'] = case['geocodes']['main']['latitude']
    dict['longitude'] = case['geocodes']['main']['longitude']
    dict['hours'] = case['hours']['display']
    dict['description'] = case["description"]

    list.append(dict)
print(list)

```

```python
df = pd.DataFrame(list)
```

```python
new_df = df.drop(df.index[[25,34,17,7,33,39,40,41,38]])
```

```python
for cafe in data['results']:
    if 'amenities' in cafe['features']:
        if 'wifi' in cafe['features']['amenities']:
            if cafe['features']['amenities']['wifi'] == 't':
                print(cafe['name'], "has wifi.")
            if cafe['features']['amenities']['wifi'] == 'f':
                print(cafe['name'], "has free wifi.")
    else:
        print("No amenities")
```

```python
for cafe in data['results']:
    if 'Open Daily 12:00 AM-11:59 PM' in cafe['hours']['display']:
        print(cafe['name'], "is open 24 hours.")
```
